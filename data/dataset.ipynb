```python
import pandas as pd
import random

import sys
!{sys.executable} -m pip install faker
from faker import Faker
from datetime import datetime

fake = Faker('en_IN')
random.seed(42)

# --------------------------
# Dimension Tables
# --------------------------

regions = pd.DataFrame({
    'RegionID':[1,2,3,4,5,6,7,8],
    'CityRegion':['Delhi','Mumbai','Bengaluru','Pune','Chennai','Kolkata','Hyderabad','Jaipur']
})

channels = pd.DataFrame({
    'ChannelID':[1,2,3],
    'PurchaseChannel':['Online','Offline','App']
})

segments = pd.DataFrame({
    'SegmentID':[1,2,3],
    'CustomerSegment':['High Value','Regular','Low Value']
})

loyalty = pd.DataFrame({
    'LoyaltyID':[1,2,3,4],
    'LoyaltyTier':['Bronze','Silver','Gold','Platinum']
})

retention = pd.DataFrame({
    'RetentionStatusID':[1,2,3],
    'RetentionStatus':['Retained','At Risk','Churned']
})

# --------------------------
# Customers
# --------------------------

customers = []
facts = []

for i in range(1,201):

    age = random.randint(21,60)
    gender = random.choice(['Male','Female'])
    occupation = random.choice(['Engineer','Teacher','Doctor','Analyst','Manager','Student'])
    income = random.choice(['Low','Medium','High'])
    cust_type = random.choice(['New','Returning'])

    customers.append([
        i,
        fake.name(),
        age,
        gender,
        occupation,
        income,
        cust_type
    ])

    region_id = random.randint(1,8)
    channel_id = random.randint(1,3)

    freq = random.randint(1,12)
    aov = random.randint(500,5000)
    spend = freq * aov

    engagement = random.randint(30,100)
    satisfaction = random.randint(1,10)

    if spend > 40000:
        segment_id = 1
        loyalty_id = 4
        retention_id = 1
    elif spend > 15000:
        segment_id = 2
        loyalty_id = 3
        retention_id = 1
    elif freq <= 2:
        segment_id = 3
        loyalty_id = 1
        retention_id = 2
    else:
        segment_id = 2
        loyalty_id = 2
        retention_id = random.choice([2,3])

    join_date = fake.date_between(start_date='-3y', end_date='-6m')
    last_purchase = fake.date_between(start_date=join_date, end_date='today')

    facts.append([
        i,
        str(join_date),
        str(last_purchase),
        region_id,
        channel_id,
        segment_id,
        loyalty_id,
        retention_id,
        freq,
        aov,
        spend,
        engagement,
        satisfaction
    ])

# Create DataFrames
dim_customer = pd.DataFrame(customers, columns=[
'CustomerID','CustomerName','Age','Gender','Occupation','IncomeLevel','CustomerType'
])

fact = pd.DataFrame(facts, columns=[
'CustomerID','JoinDate','LastPurchaseDate','RegionID','ChannelID',
'SegmentID','LoyaltyID','RetentionStatusID',
'PurchaseFrequency','AverageOrderValue','TotalSpend',
'EngagementScore','SatisfactionScore'
])

# Date Table
dates = pd.date_range(start='2023-01-01', end='2026-12-31')
dim_date = pd.DataFrame({'Date':dates})
dim_date['Year'] = dim_date['Date'].dt.year
dim_date['MonthNo'] = dim_date['Date'].dt.month
dim_date['Month'] = dim_date['Date'].dt.strftime('%B')
dim_date['Quarter'] = 'Q' + dim_date['Date'].dt.quarter.astype(str)

# Export CSVs
dim_customer.to_csv('D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Customer Segmentation & Retention Analytics in Power BI/DimCustomer.csv', index=False)
regions.to_csv('D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Customer Segmentation & Retention Analytics in Power BI/DimRegion.csv', index=False)
channels.to_csv('D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Customer Segmentation & Retention Analytics in Power BI/DimChannel.csv', index=False)
segments.to_csv('D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Customer Segmentation & Retention Analytics in Power BI/DimSegment.csv', index=False)
loyalty.to_csv('D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Customer Segmentation & Retention Analytics in Power BI/DimLoyalty.csv', index=False)
retention.to_csv('D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Customer Segmentation & Retention Analytics in Power BI/DimRetentionStatus.csv', index=False)
dim_date.to_csv('D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Customer Segmentation & Retention Analytics in Power BI/DimDate.csv', index=False)
fact.to_csv('D:/Documents/MBA/IIP Skills updrade courses/Projects/Power BI/Customer Segmentation & Retention Analytics in Power BI/FactCustomerTransactions.csv', index=False)

print("All CSV Files Created Successfully")
```
